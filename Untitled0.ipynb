```python
import pandas as pd 
import statistics 
import plotly.express as px

from google.colab import files
data_toload=files.upload()
df=pd.read_csv("savings_data_final.csv")
print(df.head())
fig=px.scatter(df,y="quant_saved", color="rem_any")
fig.show()
```

```python
import csv
with open("savings_data_final.csv", newline="") as f:
  reader=csv.reader(f)
  savings_data=list(reader)
savings_data.pop(0)
print(savings_data)
totalentries=len(savings_data)
print(totalentries)
totalnumberofpeoplegivenreminder=0
for data in savings_data:
  if int(data[3])==1:
    totalnumberofpeoplegivenreminder+=1
import plotly.graph_objects as go
fig=go.Figure(go.Bar(x=["reminded","not reminded"],y=[totalnumberofpeoplegivenreminder,(totalentries-totalnumberofpeoplegivenreminder)]))
fig.show()
```

```python
all_savings=[]
for data in savings_data:
  all_savings.append(float(data[0]))
  #print(all_savings)
print(f"Mean of savings - {statistics.mean(all_savings)}")
print(f"Median of savings - {statistics.median(all_savings)}")
print(f"Mode of savings - {statistics.mode(all_savings)}")

```

```python
import numpy as np
age=[]
savings=[]
for data in savings_data:
  if float(data[5])!=0:
    age.append(float(data[5]))
    savings.append(float(data[0]))
corelation=np.corrcoef(age,savings)
print("corelation between the age and the person and there savings is", corelation[0,1])
```

```python
import plotly.figure_factory as ff
fig=ff.create_distplot([df["quant_saved"].tolist()],["savings"],show_hist=False)
fig.show()
```

it definitely does not follow a normal distribution. If we look at this data we can see that majority of the data lies under 1000. The rest is just a small number of data points.
Because of this our chart is skewed to the left side. This means that the majority of the data instead of lying at the centre it is lying on the left side of the chart.
To deal with this problem you have to remove the outliners. We will remove the outliners with IQR(inter quartile range) method.

```python
import seaborn as sns
sns.boxplot(data=df,x=df["quant_saved"])
```

Here we can see that majority of the data is represented by a thick, long and black line. We then have a thin long black line and finally we have a horizontal line which is rest of the data.
Here thin black vertical line is the line that separates the data which we should use for analysis.

```python
q1=df["quant_saved"].quantile(0.25)
q3=df["quant_saved"].quantile(0.75)
iqr=q3-q1
print(f"q1-{q1}")
print(f"q3-{q3}")
print(f"iqr-{iqr}")
lower_whisker=q1-1.5*iqr
upper_whisker=q3+1.5*iqr
print(f"Lower Whisker - {lower_whisker}")
print(f"Upper Whisker - {upper_whisker}")

newdataframe=df[df["quant_saved"]<upper_whisker]
print(newdataframe)
```
